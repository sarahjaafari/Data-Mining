```python
#QUESTION 1
import pandas as pd
retail=pd.read_excel("Online Retail.xlsx")
retail.head()
```

```python
cleaned_retail=retail.dropna(subset=['CustomerID','StockCode'])
cleaned_retail= cleaned_retail[cleaned_retail['Quantity']>0]
nonproduct=["BANK CHARGES","C2", "DOT","M","PADS", "POST"]
cleaned_retail=cleaned_retail[~cleaned_retail['StockCode'].astype(str).isin(nonproduct)]
#display(cleaned_retail)

transactional=cleaned_retail.groupby('InvoiceNo')['StockCode'].apply(set).reset_index()
transactional.set_index('InvoiceNo',inplace=True)
binary_matrix=transactional.explode('StockCode').reset_index()
binary_matrix['Presence']= 1
binary_matrix=binary_matrix.pivot(index='InvoiceNo',columns='StockCode',values='Presence').fillna(0)
binary_matrix=binary_matrix.astype(int)
display(binary_matrix)
```

```python
from itertools import combinations

transactions=[set(binary_matrix.columns[binary_matrix.iloc[i]==1]) for i in range(len(binary_matrix))]
min_sup =0.02
num_transactions=len(transactions)
def calculate_support(itemset, transactions):
    count=sum(1 for transaction in transactions if itemset.issubset(transaction))
    return count/num_transactions
item_counts={}
for transaction in transactions:
    for item in transaction:
        item_counts[item]=item_counts.get(item,0)+1
frequent_itemsets={frozenset([item]) for item, count in item_counts.items() if count/num_transactions>=min_sup}
all_frequent_itemsets={1:frequent_itemsets}
def generate_candidates(prev_frequent,k):
    candidates=set()
    prev_frequent_list=list(prev_frequent)
    for i in range(len(prev_frequent_list)):
        for j in range(i+1,len(prev_frequent_list)):
            union_set=prev_frequent_list[i]|prev_frequent_list[j]
            if len(union_set)== k:
                candidates.add(union_set)
    return candidates
k=2
while True:
    candidates=generate_candidates(frequent_itemsets,k)
    frequent_itemsets= {itemset for itemset in candidates if calculate_support(itemset,transactions)>=min_sup}
    if not frequent_itemsets:
        break
    all_frequent_itemsets[k]=frequent_itemsets
    k+=1
for k, itemsets in all_frequent_itemsets.items():
    print(f"number of frequent {k}-itemsets: {len(itemsets)}")
for k, itemsets in all_frequent_itemsets.items():
    print(f"\nfrequent {k}-itemsets:")
    for itemset in itemsets:
        print(itemset)
```

```python
from itertools import permutations
min_conf=0.5
association_rules=[]

for k, itemsets in all_frequent_itemsets.items():
    if k<2:
        continue
    for itemset in itemsets:
        for antecedent_size in range(1,k):
            for antecedent in combinations(itemset,antecedent_size):
                antecedent=frozenset(antecedent)
                consequent=itemset-antecedent
                support_itemset=calculate_support(itemset,transactions)
                support_antecedent =calculate_support(antecedent,transactions)
                confidence=support_itemset/support_antecedent if support_antecedent>0 else 0
                if confidence>=min_conf:
                    association_rules.append((antecedent,consequent,support_itemset,confidence))

rules_df=pd.DataFrame(association_rules,columns=['Antecedent','consequent','support','Confidence'])
rules_df=rules_df.sort_values(by='Confidence',ascending=False)
print(len(rules_df)) #total number of association rules
rules_df.head(5) #top 5 association ru,es
```

```python
#QUESTION 2
column_names=["user_id","movie_id","rating","timestamp"]
ratings=pd.read_csv("ml-100k/u.data",sep="\t",names=column_names,usecols=["user_id","movie_id","rating", "timestamp"])
ratings.head()
```

```python
user_counts=ratings["user_id"].value_counts()
valid_users=user_counts[user_counts>=20].index
filtered_ratings=ratings[ratings["user_id"].isin(valid_users)]

movie_counts=filtered_ratings["movie_id"].value_counts()
valid_movies= movie_counts[movie_counts>=20].index
filtered_ratings=filtered_ratings[filtered_ratings["movie_id"].isin(valid_movies)]

filtered_ratings.head()
```

```python
ratings=ratings.sort_values(by="timestamp")
split_index=int(len(ratings)*0.8)
train_data=ratings.iloc[:split_index]
test_data=ratings.iloc[split_index:]
train_users=set(train_data["user_id"])
test_data=test_data[test_data["user_id"].isin(train_users)]

train_movies=set(train_data["movie_id"])
test_data=test_data[test_data["movie_id"].isin(train_movies)]
```

```python
train_data.to_csv("train_data.csv",index=False) #saving it in csv for easier use
test_data.to_csv("test_data.csv",index=False) #...
```

```python
import numpy as np
train_data=pd.read_csv("train_data.csv")
user_item_matrix=train_data.pivot(index="user_id",columns="movie_id",values="rating").fillna(0)
matrix=user_item_matrix.values
U, sigma, Vt=np.linalg.svd(matrix,full_matrices=False)
sigma=np.diag(sigma)

def reconstruct_matrix(k):
    return np.dot(np.dot(U[:, :k], sigma[:k, :k]), Vt[:k, :])

k_values=[5,10,20,40]
reconstructed_matrices={k: reconstruct_matrix(k) for k in k_values}

predicted_matrix=pd.DataFrame(reconstructed_matrices[10],index=user_item_matrix.index, columns=user_item_matrix.columns)
def predict_rating(user_id,movie_id):
    return predicted_matrix.loc[user_id,movie_id] if movie_id in predicted_matrix.columns and user_id in predicted_matrix.index else np.nan

print(f"predicted rating: {predict_rating(5, 100):.2f}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

test_data=pd.read_csv("test_data.csv")

def evaluate_model(k):
    reconstructed=pd.DataFrame(reconstructed_matrices[k],index=user_item_matrix.index, columns=user_item_matrix.columns)
    test_data_filtered=test_data[test_data["user_id"].isin(reconstructed.index) & test_data["movie_id"].isin(reconstructed.columns)]
    true_ratings=test_data_filtered["rating"].values
    predicted_ratings=[reconstructed.loc[u, m] for u, m in zip(test_data_filtered["user_id"],test_data_filtered["movie_id"])]
    
    rmse=np.sqrt(mean_squared_error(true_ratings,predicted_ratings))
    mae=mean_absolute_error(true_ratings,predicted_ratings)
    return rmse, mae

k_values=[5,10,20,40]
results={k: evaluate_model(k) for k in k_values}

results_df=pd.DataFrame(results,index=["RMSE", "MAE"]).T
print(results_df)
```

Impact of k on model performance
A lower k value leads to high reconstruction error due to excessive information loss, while a higher k captures more latent factors, reducing RMSE and MAE. However, an excessively high k can cause overfitting, where the model memorizes data rather than generalizing well. The optimal k balances accuracy and generalization.

Strengths/limitationd of matrix factorization
Matrix factorization captures hidden patterns in user preferences, handles sparse data well, and scales efficiently for large datasets, making it ideal for recommendation systems. Limitations include difficulty recommending for new users/items, requires careful tuning of k, and can be computationally expensive for very large datasets.
